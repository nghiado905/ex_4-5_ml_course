```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(0, 10, 100)  
epsilon = np.random.normal(0, np.sqrt(4), size=len(x))  
y = 2 * x + epsilon

x = x.reshape(-1, 1)
```

Linear Regression model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)

```

Poly Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Polynomial regression (9th degree)
poly_model = make_pipeline(PolynomialFeatures(9), LinearRegression())
poly_model.fit(x, y)
```

```python
y_pred_linear = model.predict(x)
y_pred_poly = poly_model.predict(x)
```

Compute metrics to compare

```python
from sklearn.metrics import mean_squared_error

mse_linear = mean_squared_error(y_pred_linear,y)
mse_poly = mean_squared_error(y_pred_poly,y)


print(f'MSE of Linear {mse_linear:.2f}')
print(f'MSE of Poly {mse_poly:.2f}')
```

PLOT OF LINEAR REGRESSION

```python
# Plot data
plt.scatter(x, y, color='blue', label='Data', alpha=0.6)
plt.plot(x, y_pred_linear, color='red', label='Linear Regression')

```

PLOT OF POLYNOMINAL REGRESSION

```python
plt.scatter(x, y, color='blue', label='Data', alpha=0.6)
plt.plot(x, y_pred_poly, color='green', label='Poly Regresson')

```
